### SETUP

```python
import pandas as pd
```

```python
vg_df = pd.read_csv('./assets/vgsales.csv')
vg_df
```

#### Q1: Which company is the most common video game publisher?

```python
top_publisher = vg_df['Publisher'].mode()[0]
top_publisher
```

#### Q2: Whatâ€™s the most common platform?

```python
top_platform = vg_df['Platform'].mode()[0]
top_platform
```

#### Q3: What about the most common genre?

```python
top_genre = vg_df['Genre'].mode()[0]
top_genre
```

#### Q4: What are the top 20 highest grossing games?

```python
top_20_grossing = vg_df[['Name', 'Global_Sales']].nlargest(20, ['Global_Sales'])
top_20_grossing
```

#### Q5: For North American video game sales, whatâ€™s the median?

```python
na_median_sales = vg_df['NA_Sales'].median()
na_median_sales
```

#### Q5a - Provide a secondary output showing ten games surrounding the median sales output
- assume that games with same median value are sorted in descending order

```python
na_median_subset = vg_df.loc[vg_df['NA_Sales'] == na_median_sales]
na_sales_median_middle = len(na_median_subset) // 2
na_median_games = na_median_subset[na_sales_median_middle-5:na_sales_median_middle+6]
na_median_games
```
